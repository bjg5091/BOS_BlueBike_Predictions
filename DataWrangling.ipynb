# 2. BlueBike station information

```python
# Call the API and pull out data into the JSON structure that will be returned
r=requests.get("https://member.bluebikes.com/data/stations.json")
bike_data=r.json()
```

```python
print(bike_data.keys())
station=json_normalize(bike_data["stations"])
station.head(10)
```

#### We successfully pulled the real-time station information, however it is not clear what each column represents. 
#### Fortunately, the real-time data is also available in xml format at "https://member.bluebikes.com/data/stations/bikeStations.xml"

#### By reading the xml feed, we deduce what each column in station dataframe represents. 

#### In particular we will take from each station the following data: 

1. station id
2. station name
3. latitude, longitude
4. district (municipality)
5. the number of bikes available
6. empty dock available


```python
# Take station id, station name, latitude, longitude, district(municipality), bike available, (empty) dock available
station=station[["id", "s", "la", "lo", "d", "ba","da"]]
```

```python
# Bike available + (emtpy) dock available  = total number of docks at the station
station["#_of_Docks"] = station.ba + station.da

# Drop ba, da
station=station.drop(["ba", "da"], axis=1)

station.head()
```

```python
# Update column names to identifable words
station.columns=["ID", "Station", "Latitude", "Longitude", "Municipality", "#_of_Docks"]
```

# 3. Merging trip history data and station data

#### Recall that there are 321 unique station id's in trip_simp where as there are only 277 unique station id's in station. 

#### We choose outer merge since we do not know that the set unique id's in one dataset is a proper subset of that in another.

```python
# Merge trip_simp and station to see the matches
trip_endinfo=pd.merge(trip_simp, station, left_on="end_station_id", right_on="ID", how="outer")
trip_endinfo.head(5)
```

```python
# We find station id's in tripdf that did not get matched 
unmatched=trip_endinfo[trip_endinfo.ID.isnull()].groupby(["end_station_id"]).first()
unmatched.head(10)
```

#### There are total 50 unmatched stations

```python
len(unmatched)
```

#### We create a list of unmatched id's in the trip_endinfo.

```python
fixidlist=unmatched.index.values
fixidlist = fixidlist.astype(int)
fixidlist
```

#### We manually checked on BlueBike website to see where these stations are located. In particular, we checked if the unmatched stations can be identified. We found that 
1. Some stations were relocated, but very little. For example, they were moved diagonally in the same intersection.  The original station location was listed under one id, and the new station location was listed under the id. In this case, we will map the old station id's to new station id's. 
2. Some stations are test stations. We will eventually remove any trip observations that start from or end at these test stations. 

```python
# Construct a dictionay based on the findings
fixdict = {1:np.nan, 13:385, 88:381, 155:331, 158:np.nan, 164:np.nan, 168:332, 220:328, 229:np.nan,
           230:np.nan, 231:329, 237:333, 238:334, 240:335, 241:336, 242:337, 243:338, 244:339, 
           245:340, 246:341, 247:342, 248:343, 249:344,250:345, 252:346, 253:347, 254:348, 256:349, 
           257:350, 262:351, 263:353, 264:354, 274:355, 277:356, 278:357, 281:358, 283:327, 284:359, 
           286:330, 293:360, 294:361, 295:362, 299:363, 300:364, 302:365, 306:368, 307:366, 308:np.nan, 
           311:367, 317:np.nan}
```

```python
# Fix the id's
trip_simp=trip_simp.replace({"start_station_id":fixdict})
trip_simp=trip_simp.replace({"end_station_id":fixdict})
```

```python
# Drop any trip that involves null values. These trips are from and to test stations. 

trip_simp=trip_simp.dropna(subset=["start_station_id"])
trip_simp=trip_simp.dropna(subset=["end_station_id"])
```

```python
# Check the dataset
trip_simp.info()
```

```python
# We want station id's as int not float
trip_simp.start_station_id = trip_simp.start_station_id.astype(int)
trip_simp.end_station_id = trip_simp.end_station_id.astype(int)
```

# 4. EDA on Trip history dataset 

```python
# Check the distribution of start station ane end station id's
fig = plt.figure()

plt.subplot(1,2,1)
trip_simp.start_station_id.plot(kind="hist", bins=384)

plt.subplot(1,2,2)
trip_simp.end_station_id.plot(kind="hist", bins=384)

plt.tight_layout()
```

```python
# Find popular start stations. Store (station id, station name, the number of occurance) in a list
pop_starts=trip_simp.start_station_id.value_counts()
startslist=[]
for (start_id, value) in pop_starts.iteritems():
    startslist.append((start_id, station.loc[station.ID == start_id]["Station"].values[0],value))
```

```python
# Find popular end stations. Store (station id, station name, the number of occurance) in a list
pop_ends=trip_simp.end_station_id.value_counts()
endslist=[]
for (end_id, value) in pop_ends.iteritems():
    endslist.append((end_id, station.loc[station.ID == end_id]["Station"].values[0],value))
```

```python
# Convert the lists above to DataFrames
pop_starts_df=pd.DataFrame(startslist)
pop_starts_df.columns=["start_station_id", "start_station_name", "count"]

pop_ends_df=pd.DataFrame(endslist)
pop_ends_df.columns=["end_station_id", "end_station_name", "count"]
```

```python
pop_starts_df.head()
```

```python
pop_ends_df.head()
```

#### We plot the location of all BlueBike stations on top of layers of MBTA (Boston subway system) lines and MBTA stations. 

```python
# The coordinate reference system using lat/long is epsg:4326
crs= {'init':'epsg:4326'}

# Create a point object using latitude, longitude. Notice the change of order. 
geometry= [Point(xy) for xy in zip(station.Longitude, station.Latitude)]

# Define a new dataframe with this point object
station_geo= gpd.GeoDataFrame(station, crs=crs, geometry = geometry)

# The MBTA (metro system) shape files from Massachusetts government follow different coordinate reference system. 
# We therefore transform the geolocation to fit the shapefiles
station_geo_trans = station_geo.to_crs(epsg=26986)

# Read the shapefiles
metroarc= gpd.read_file("mbta_rapid_transit/MBTA_ARC.shp")
metronode= gpd.read_file("mbta_rapid_transit/MBTA_NODE.shp")

# Customize the size and axes
fig, ax=plt.subplots(figsize=(10,10))
ax.set_xlim([225000, 245000])
ax.set_ylim([890000, 910000])

# Plot metro line in grey
metroarc.plot(ax=ax, alpha=0.4, color="grey")

# Plot metro stations in small black dots
metronode.plot(ax=ax, color="k",  markersize=4)

# Plot the bike stations in blue
station_geo_trans.plot(ax=ax, color="blue", alpha=0.4)
plt.show()
```

#### We improve the above plot by incorporating the popularity of each bike station. The markersize denotes the popularity of bike station. We produce two figures--one for the popularity of start stations and another for the end stations. 

```python
# The start stations are plotted in green
crs={'init':'epsg:4326'}
pop_starts_loc=pd.merge(pop_starts_df, station[["ID", "Latitude", "Longitude"]], left_on="start_station_id", right_on="ID", how="left")
geometry=[Point(xy) for xy in zip(pop_starts_loc.Longitude, pop_starts_loc.Latitude)]
pop_starts_geo=gpd.GeoDataFrame(pop_starts_loc, crs=crs, geometry = geometry).to_crs(epsg=26986)
fig, ax=plt.subplots(figsize=(10,10))
ax.set_xlim([225000, 242500])
ax.set_ylim([890000, 907500])
metroarc.plot(ax=ax, alpha=0.4, color="grey")
metronode.plot(ax=ax, alpha=0.4, color="blue",  markersize=4)
pop_starts_geo.plot(ax=ax, color="green", alpha=0.8, markersize=[math.pow(n, 0.5) for n in pop_starts_geo["count"]])

# The end stations are plotted in brown
crs={'init':'epsg:4326'}
pop_ends_loc=pd.merge(pop_ends_df, station[["ID", "Latitude", "Longitude"]], left_on="end_station_id", right_on="ID", how="left")
geometry=[Point(xy) for xy in zip(pop_ends_loc.Longitude, pop_ends_loc.Latitude)]
pop_ends_geo=gpd.GeoDataFrame(pop_ends_loc, crs=crs, geometry = geometry).to_crs(epsg=26986)
fig, ax=plt.subplots(figsize=(10, 10))
ax.set_xlim([225000, 242500])
ax.set_ylim([890000, 907500])
metroarc.plot(ax=ax, alpha=0.4, color="grey")
metronode.plot(ax=ax, alpha=0.4, color="blue",  markersize=4)
pop_ends_geo.plot(ax=ax, color="brown", alpha=0.8, markersize=[math.pow(n, 0.5) for n in pop_ends_geo["count"]])
```

# 5. Daily weather data

```python
# Read daily weather data from https://www.ncdc.noaa.gov/cdo-web/datasets#GHCND
daily=pd.read_csv("DailyWeather.csv", low_memory=False, parse_dates=["DATE"])
daily=daily.set_index("DATE")
```

```python
daily.notnull().sum()
```

#### There are 136 observatories near Boston, MA. We will only use data under "Boston, MA US"

```python
# Plot daily average temperature
dailyBoston=daily[daily.NAME=="BOSTON, MA US"]
dailyBoston.TAVG.plot()
```

#### We explore the correlation between the daily temperature and daily bike trips. 

```python
# Use starttime as index in trip_simp
trip_time=trip_simp.set_index("starttime")

# Resample to get daily counts
tripperday=trip_time.resample("D").count()
tripperday.head()
```

```python
# Plot two series: the daily average temperature, the number of trips per day
fig, ax1 = plt.subplots(figsize=(10, 7))
ax1.plot(dailyBoston.TAVG)

ax2 = ax1.twinx()  # instantiate a second axis that shares the same x-axis
ax2.plot(tripperday.tripduration, color="red")

fig.tight_layout() 
plt.show()
```

#### What if we take rolling means? 

```python
fig, ax1 = plt.subplots(figsize=(10, 7))
ax1.plot(dailyBoston.TAVG.rolling(7).mean())

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
ax2.plot(tripperday.tripduration.rolling(7).mean(), color="red")

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
